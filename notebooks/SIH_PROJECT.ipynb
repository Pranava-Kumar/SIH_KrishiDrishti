```python
import torch
from torch.utils.data import Dataset
import numpy as np

# Parameters
patch_size = 64
stride = 32

H, W, B = hs_cube.shape
patches = []
labels = []

for i in range(0, H - patch_size + 1, stride):
    for j in range(0, W - patch_size + 1, stride):
        patch = hs_cube[i:i+patch_size, j:j+patch_size, :]
        label_patch = gt_cube[i:i+patch_size, j:j+patch_size]

        # majority label, skip background 0 if needed
        counts = np.bincount(label_patch.flatten().astype(int))
        if len(counts) <= 1:  # all background
            continue
        label = np.argmax(counts[1:]) + 1  # skip 0
        patches.append(patch)
        labels.append(label)

# Convert to NumPy arrays
patches = np.stack(patches, axis=0)          # (N, patch_size, patch_size, Bands)
patches = patches.transpose(0, 3, 1, 2)      # (N, Bands, H, W)

# Normalize HS patches to 0-1
patches = (patches - patches.min()) / (patches.max() - patches.min())

print("Prepared patches shape:", patches.shape)
print("Number of patches:", len(labels))
```

```python
class HonghuPatchDataset(Dataset):
    def __init__(self, patches, labels):
        self.patches = torch.tensor(patches, dtype=torch.float32)
        self.labels = torch.tensor(labels, dtype=torch.long)

    def __len__(self):
        return len(self.labels)

    def __getitem__(self, idx):
        return self.patches[idx], self.labels[idx]

hs_dataset = HonghuPatchDataset(patches, labels)
```

```python
# =========================================
# Production-Ready Combined Hyperspectral + RGB Model
# =========================================

import os
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, TensorDataset
from torchvision import datasets, transforms, models
import spectral as spy
import numpy as np
from PIL import Image
from sklearn.preprocessing import LabelEncoder
from tqdm import tqdm

# -------------------------
# Device setup
# -------------------------
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

# -------------------------
# Hyperparameters
# -------------------------
BATCH_SIZE = 16
EPOCHS = 20
LR = 1e-3
HS_CUBE_SIZE = (64, 64)  # resize HS cubes
RGB_IMG_SIZE = (224, 224)

# -------------------------
# Load HongHu Hyperspectral Dataset
# -------------------------
honghu_path = "/root/.cache/huggingface/hub/datasets--danaroth--whu_hi/snapshots/480e209046fe2711463113cf9aff1f0d1c5a4406/WHU-Hi-HongHu"
hdr_file = os.path.join(honghu_path, "WHU-Hi-HongHu.hdr")
img_hs = spy.open_image(hdr_file)
cube = img_hs.load()  # shape: (H, W, Bands)

# Load ground truth labels
gt_file = os.path.join(honghu_path, "WHU-Hi-HongHu_gt.hdr")
gt_img = spy.open_image(gt_file)
gt = gt_img.load()  # shape: (H, W)

# Preprocess HS dataset: Extract small patches
def extract_patches(cube, gt, patch_size=HS_CUBE_SIZE[0]):
    H, W, Bands = cube.shape
    patches = []
    labels = []
    stride = patch_size
    for i in range(0, H - patch_size + 1, stride):
        for j in range(0, W - patch_size + 1, stride):
            patch = cube[i:i+patch_size, j:j+patch_size, :]
            label_patch = gt[i:i+patch_size, j:j+patch_size]
            label = np.bincount(label_patch.flatten().astype(int)).argmax()
            if label > 0:  # only keep non-background
                patches.append(np.transpose(patch, (2,0,1)))  # Bands x H x W
                labels.append(label)
    patches = np.array(patches, dtype=np.float32)
    labels = np.array(labels, dtype=np.int64)
    return patches, labels

hs_patches, hs_labels = extract_patches(cube, gt)
print("HS patches:", hs_patches.shape, "Labels:", hs_labels.shape)

hs_dataset = TensorDataset(torch.tensor(hs_patches), torch.tensor(hs_labels))
hs_loader = DataLoader(hs_dataset, batch_size=BATCH_SIZE, shuffle=True, drop_last=True) # Added drop_last=True

# -------------------------
# Load PlantVillage RGB Dataset
# -------------------------
kaggle_path = "/kaggle/input/plantdisease/plantvillage/PlantVillage"

rgb_transform = transforms.Compose([
    transforms.Resize(RGB_IMG_SIZE),
    transforms.ToTensor()
])

plant_dataset = datasets.ImageFolder(root=kaggle_path, transform=rgb_transform)
plant_loader = DataLoader(plant_dataset, batch_size=BATCH_SIZE, shuffle=True, drop_last=True) # Added drop_last=True
num_rgb_classes = len(plant_dataset.classes)
print("Number of RGB classes:", num_rgb_classes)

# Encode labels for RGB dataset
rgb_label_encoder = LabelEncoder()
rgb_labels_encoded = rgb_label_encoder.fit_transform([y for _, y in plant_dataset.samples])

# -------------------------
# Models
# -------------------------
class HyperspectralCNN(nn.Module):
    def __init__(self, input_bands, output_dim=128):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(input_bands, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Flatten()
        )
        self.fc = nn.Linear(128*(HS_CUBE_SIZE[0]//4)*(HS_CUBE_SIZE[1]//4), output_dim)
    def forward(self, x):
        x = self.features(x)
        # print("Shape before FC layer in HS_CNN:", x.shape) # Keep this for internal check if needed
        x = self.fc(x)
        return x

class RGBResNet(nn.Module):
    def __init__(self, output_dim=128):
        super().__init__()
        self.base_model = models.resnet18(pretrained=True)
        self.base_model.fc = nn.Linear(self.base_model.fc.in_features, output_dim)
    def forward(self, x):
        return self.base_model(x)

class CombinedModel(nn.Module):
    def __init__(self, hs_model, rgb_model, final_classes):
        super().__init__()
        self.hs_model = hs_model
        self.rgb_model = rgb_model
        self.classifier = nn.Sequential(
            nn.Linear(128+128, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, final_classes)
        )
    def forward(self, hs_x, rgb_x):
        hs_feat = self.hs_model(hs_x)
        rgb_feat = self.rgb_model(rgb_x)
        print("Shape of hs_feat:", hs_feat.shape) # Added print
        print("Shape of rgb_feat:", rgb_feat.shape) # Added print
        combined = torch.cat([hs_feat, rgb_feat], dim=1)
        print("Shape of combined:", combined.shape) # Added print
        out = self.classifier(combined)
        return out

# Initialize models
hs_model = HyperspectralCNN(input_bands=hs_patches.shape[1]).to(device)
rgb_model = RGBResNet().to(device)
combined_model = CombinedModel(hs_model, rgb_model, final_classes=num_rgb_classes).to(device)

# -------------------------
# Training loop
# -------------------------
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(combined_model.parameters(), lr=LR)

def train_model(combined_model, hs_loader, rgb_loader, epochs=EPOCHS):
    combined_model.train()
    for epoch in range(epochs):
        hs_iter = iter(hs_loader)
        rgb_iter = iter(rgb_loader)
        running_loss = 0.0
        # Iterate for the minimum number of batches available after dropping the last one
        for _ in range(min(len(hs_loader), len(rgb_loader))):
            hs_data, hs_label = next(hs_iter)
            rgb_data, rgb_label = next(rgb_iter)

            hs_data, hs_label = hs_data.to(device), hs_label.to(device)
            rgb_data, rgb_label = rgb_data.to(device), rgb_label.to(device)

            optimizer.zero_grad()
            outputs = combined_model(hs_data, rgb_data)
            # Use PlantVillage labels for loss
            loss = criterion(outputs, rgb_label)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
        print(f"Epoch [{epoch+1}/{epochs}] Loss: {running_loss/min(len(hs_loader), len(rgb_loader)):.4f}")
    print("Training finished.")
    return combined_model

# Train the model
best_model = train_model(combined_model, hs_loader, plant_loader, epochs=EPOCHS)

# Save model
torch.save(best_model.state_dict(), "combined_model.pth")
print("Model saved as combined_model.pth")
```

```python
import os
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, Dataset, TensorDataset
from torchvision import datasets, transforms, models
import spectral as spy
import numpy as np
from sklearn.preprocessing import LabelEncoder
from tqdm import tqdm

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)

BATCH_SIZE = 16
EPOCHS = 10
LR = 1e-3
HS_CUBE_SIZE = (64, 64)
RGB_IMG_SIZE = (224, 224)


honghu_path = "/root/.cache/huggingface/hub/datasets--danaroth--whu_hi/snapshots/480e209046fe2711463113cf9aff1f0d1c5a4406/WHU-Hi-HongHu"
hdr_file = os.path.join(honghu_path, "WHU-Hi-HongHu.hdr")
img_hs = spy.open_image(hdr_file)
cube = img_hs.load()

gt_file = os.path.join(honghu_path, "WHU-Hi-HongHu_gt.hdr")
gt_img = spy.open_image(gt_file)
gt = gt_img.load()

def extract_patches(cube, gt, patch_size=HS_CUBE_SIZE[0]):
    H, W, Bands = cube.shape
    patches = []
    labels = []
    stride = patch_size
    for i in range(0, H - patch_size + 1, stride):
        for j in range(0, W - patch_size + 1, stride):
            patch = cube[i:i+patch_size, j:j+patch_size, :]
            label_patch = gt[i:i+patch_size, j:j+patch_size]
            label = np.bincount(label_patch.flatten().astype(int)).argmax()
            if label > 0:
                patches.append(np.transpose(patch, (2,0,1)))
                labels.append(label)
    return np.array(patches, dtype=np.float32), np.array(labels, dtype=np.int64)

hs_patches, hs_labels = extract_patches(cube, gt)
print("HS patches:", hs_patches.shape, "Labels:", hs_labels.shape)

hs_dataset = TensorDataset(torch.tensor(hs_patches), torch.tensor(hs_labels))
hs_loader = DataLoader(hs_dataset, batch_size=BATCH_SIZE, shuffle=False, drop_last=True)


kaggle_path = "/kaggle/input/plantdisease/plantvillage/PlantVillage" # Replay with cuser Colab cach
rgb_transform = transforms.Compose([
    transforms.Resize(RGB_IMG_SIZE),
    transforms.ToTensor()
])
plant_dataset = datasets.ImageFolder(root=kaggle_path, transform=rgb_transform)
plant_loader = DataLoader(plant_dataset, batch_size=BATCH_SIZE, shuffle=False, drop_last=True)
num_rgb_classes = len(plant_dataset.classes)
print("Number of RGB classes:", num_rgb_classes)


rgb_label_encoder = LabelEncoder()
rgb_labels_encoded = rgb_label_encoder.fit_transform([y for _, y in plant_dataset.samples])


class HyperspectralCNN(nn.Module):
    def __init__(self, input_bands, output_dim=128):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(input_bands, 64, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(64, 128, kernel_size=3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Flatten()
        )
        self.fc = nn.Linear(128*(HS_CUBE_SIZE[0]//4)*(HS_CUBE_SIZE[1]//4), output_dim)
    def forward(self, x):
        x = self.features(x)
        x = self.fc(x)
        return x

class RGBResNet(nn.Module):
    def __init__(self, output_dim=128):
        super().__init__()
        self.base_model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)
        self.base_model.fc = nn.Linear(self.base_model.fc.in_features, output_dim)
    def forward(self, x):
        return self.base_model(x)

hs_model = HyperspectralCNN(input_bands=hs_patches.shape[1]).to(device)
hs_model.eval()
hs_features_list = []
with torch.no_grad():
    for hs_data, _ in tqdm(hs_loader, desc="Extracting HS features"):
        hs_data = hs_data.to(device)
        hs_feat = hs_model(hs_data)
        hs_features_list.append(hs_feat.cpu())

hs_features = torch.cat(hs_features_list)
torch.save(hs_features, "hs_features.pt")
print("HS features saved:", hs_features.shape)


class HS_RGB_Dataset(Dataset):
    def __init__(self, hs_features, rgb_dataset):
        self.hs_features = hs_features
        self.rgb_dataset = rgb_dataset
    def __len__(self):
        return len(self.rgb_dataset)
    def __getitem__(self, idx):
        rgb_img, rgb_label = self.rgb_dataset[idx]
        hs_feat = self.hs_features[idx % len(self.hs_features)]
        return hs_feat, rgb_img, rgb_label

combined_dataset = HS_RGB_Dataset(hs_features, plant_dataset)
combined_loader = DataLoader(combined_dataset, batch_size=BATCH_SIZE, shuffle=True, drop_last=True)


class CombinedModel(nn.Module):
    def __init__(self, rgb_model, final_classes):
        super().__init__()
        self.rgb_model = rgb_model
        self.classifier = nn.Sequential(
            nn.Linear(128+128, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, final_classes)
        )
    def forward(self, hs_feat, rgb_img):
        rgb_feat = self.rgb_model(rgb_img)
        combined = torch.cat([hs_feat, rgb_feat], dim=1)
        out = self.classifier(combined)
        return out

rgb_model = RGBResNet().to(device)
combined_model = CombinedModel(rgb_model, final_classes=num_rgb_classes).to(device)

criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(combined_model.parameters(), lr=LR)

for epoch in range(EPOCHS):
    combined_model.train()
    running_loss = 0.0
    for hs_feat, rgb_data, rgb_label in combined_loader:
        hs_feat = hs_feat.to(device)
        rgb_data, rgb_label = rgb_data.to(device), rgb_label.to(device)

        optimizer.zero_grad()
        outputs = combined_model(hs_feat, rgb_data)
        loss = criterion(outputs, rgb_label)
        loss.backward()
        optimizer.step()
        running_loss += loss.item()
    print(f"Epoch [{epoch+1}/{EPOCHS}] Loss: {running_loss/len(combined_loader):.4f}")

torch.save(combined_model.state_dict(), "combined_model.pth")
print("Combined model saved as combined_model.pth")
```

```python
#Testing
import torch
from torch.utils.data import Dataset, DataLoader
from torchvision import transforms, datasets
import spectral as spy
import numpy as np
from sklearn.preprocessing import LabelEncoder


device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Using device:", device)


BATCH_SIZE = 16
HS_CUBE_SIZE = (64, 64)
RGB_IMG_SIZE = (224, 224)
NUM_RGB_CLASSES = 15  # same as training


hs_features = torch.load("hs_features.pt")  # precomputed HS features


kaggle_path = "/kaggle/input/plantdisease/plantvillage/PlantVillage"
rgb_transform = transforms.Compose([
    transforms.Resize(RGB_IMG_SIZE),
    transforms.ToTensor()
])
plant_dataset = datasets.ImageFolder(root=kaggle_path, transform=rgb_transform)


rgb_label_encoder = LabelEncoder()
rgb_labels_encoded = rgb_label_encoder.fit_transform([y for _, y in plant_dataset.samples])
class_names = plant_dataset.classes


class HS_RGB_TestDataset(Dataset):
    def __init__(self, hs_features, rgb_dataset):
        self.hs_features = hs_features
        self.rgb_dataset = rgb_dataset
    def __len__(self):
        return len(self.rgb_dataset)
    def __getitem__(self, idx):
        rgb_img, rgb_label = self.rgb_dataset[idx]
        hs_feat = self.hs_features[idx % len(self.hs_features)]
        return hs_feat, rgb_img, rgb_label

test_dataset = HS_RGB_TestDataset(hs_features, plant_dataset)
test_loader = DataLoader(test_dataset, batch_size=BATCH_SIZE, shuffle=False)


import torch.nn as nn
from torchvision import models

class HyperspectralCNN(nn.Module):
    def __init__(self, input_bands, output_dim=128):
        super().__init__()
        self.features = nn.Sequential(
            nn.Conv2d(input_bands, 64, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Conv2d(64, 128, 3, padding=1),
            nn.ReLU(),
            nn.MaxPool2d(2),
            nn.Flatten()
        )
        self.fc = nn.Linear(128*(HS_CUBE_SIZE[0]//4)*(HS_CUBE_SIZE[1]//4), output_dim)
    def forward(self, x):
        x = self.features(x)
        x = self.fc(x)
        return x

class RGBResNet(nn.Module):
    def __init__(self, output_dim=128):
        super().__init__()
        self.base_model = models.resnet18(weights=models.ResNet18_Weights.DEFAULT)
        self.base_model.fc = nn.Linear(self.base_model.fc.in_features, output_dim)
    def forward(self, x):
        return self.base_model(x)

class CombinedModel(nn.Module):
    def __init__(self, rgb_model, final_classes):
        super().__init__()
        self.rgb_model = rgb_model
        self.classifier = nn.Sequential(
            nn.Linear(128+128, 256),
            nn.ReLU(),
            nn.Dropout(0.3),
            nn.Linear(256, final_classes)
        )
    def forward(self, hs_feat, rgb_img):
        rgb_feat = self.rgb_model(rgb_img)
        combined = torch.cat([hs_feat, rgb_feat], dim=1)
        out = self.classifier(combined)
        return out

rgb_model = RGBResNet().to(device)
combined_model = CombinedModel(rgb_model, final_classes=NUM_RGB_CLASSES).to(device)
combined_model.load_state_dict(torch.load("combined_model.pth", map_location=device))
combined_model.eval()

all_predictions = []
all_labels = []

with torch.no_grad():
    for hs_feat, rgb_data, labels in test_loader:
        hs_feat = hs_feat.to(device)
        rgb_data, labels = rgb_data.to(device), labels.to(device)

        outputs = combined_model(hs_feat, rgb_data)
        _, preds = torch.max(outputs, 1)

        all_predictions.extend(preds.cpu().numpy())
        all_labels.extend(labels.cpu().numpy())


accuracy = np.mean(np.array(all_predictions) == np.array(all_labels))
print(f"Test Accuracy: {accuracy*100:.2f}%")


for i, pred in enumerate(all_predictions[:20]):  # first 20 samples
    print(f"Sample {i}: Predicted={class_names[pred]}, True={class_names[all_labels[i]]}")
```
